```python
#Task 11 Group by region and find out which region is more profitable
```

```python
profit_region=order_df.groupby(['Region'])['Profit'].sum().to_frame(name='Total Profit').sort_values(by='Total Profit',ascending = False).reset_index()
```

```python
profit_region
```

```python
Total_Orders = order_df.groupby(by=['Region'])['Order ID'].count().to_frame(name='Total Orders').sort_values(by='Total Orders',ascending = False).reset_index()
```

```python
Total_Orders
```

```python
temp_merged=profit_region.merge(Total_Orders)
```

```python
temp_merged['Avg_Profit']=temp_merged['Total Profit']/temp_merged['Total Orders']
```

```python
temp_merged   ##based on obsrevation, Central region is profitable
```
